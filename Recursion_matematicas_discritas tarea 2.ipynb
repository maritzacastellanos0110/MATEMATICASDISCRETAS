```python
# Este código demuestra el concepto de recursión mediante dos ejemplos:
# 1. Secuencia de Fibonacci: un ejemplo matemático donde cada número es la suma de los dos anteriores.
# 2. Árbol fractal: un ejemplo de recursión visual que representa la autosimilitud en el arte, donde cada rama genera dos ramas más pequeñas.

import matplotlib.pyplot as plt
import numpy as np

# 1. Secuencia de Fibonacci (Recursión Matemática)
def fibonacci(n, sequence=[1, 1]):
    if len(sequence) >= n:
        return sequence[:n]
    else:
        sequence.append(sequence[-1] + sequence[-2])
        return fibonacci(n, sequence)

# Generar y graficar los primeros 10 números de Fibonacci
fib_sequence = fibonacci(10)

# Crear una figura con un fondo suave
plt.figure(figsize=(12, 6), facecolor='lightpink')

# Graficar la secuencia de Fibonacci
plt.subplot(1, 2, 1)
plt.plot(range(1, 11), fib_sequence, 'o-', color='salmon', markersize=10, linewidth=2, label='Fibonacci Sequence')
plt.fill_between(range(1, 11), fib_sequence, color='lightcoral', alpha=0.5)  # Área bajo la curva
plt.xlabel('n', fontsize=14)
plt.ylabel('Fibonacci(n)', fontsize=14)
plt.title('Secuencia de Fibonacci (Recursión Matemática)', fontsize=16, color='darkred')
plt.xticks(range(1, 11), fontsize=12)
plt.yticks(fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)

# 2. Fractal de Árbol (Recursión en el Arte)
def draw_tree(x, y, angle, depth, length=100):
    if depth > 0:
        # Calcular nuevas coordenadas del segmento
        x_new = x + int(np.cos(np.radians(angle)) * length * depth / 10)
        y_new = y + int(np.sin(np.radians(angle)) * length * depth / 10)
        plt.plot([x, x_new], [y, y_new], color=plt.cm.Paired(depth / 7), lw=depth, alpha=0.7)  # Dibuja la rama
        # Llamada recursiva para ramas izquierda y derecha
        draw_tree(x_new, y_new, angle - 30, depth - 1, length)  # Rama izquierda
        draw_tree(x_new, y_new, angle + 30, depth - 1, length)  # Rama derecha

# Graficar el árbol fractal
plt.subplot(1, 2, 2)
plt.title('Árbol Fractal (Recursión en el Arte)', fontsize=16, color='darkgreen')
draw_tree(0, -150, 90, 7, 100)
plt.axis('off')  # Ocultar los ejes
plt.gca().set_facecolor('lavender')  # Cambiar el fondo del gráfico a un color suave

# Mostrar gráficos
plt.tight_layout()
plt.show()




```
